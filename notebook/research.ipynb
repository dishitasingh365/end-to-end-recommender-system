### kaggle dataset link - https://www.kaggle.com/datasets/ra4u12/bookrecommendation

```python
#importing BX-Books.csv
books = pd.read_csv(
    '/Users/dishitasingh/Desktop/PersonalProjects/end-to-end-recommender-system/notebook/BX-Books.csv',
    sep=';',
    encoding='latin-1',
    on_bad_lines='skip'
)
```

```python
books.head()
```

```python
#taking only needed columns
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
# Lets remane some wierd columns name
books.rename(columns={
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": "publisher",
    "Image-URL-L": "image_url"
}, inplace=True)
```

```python
#importing BX-Users.csv
users = pd.read_csv(
    '/Users/dishitasingh/Desktop/PersonalProjects/end-to-end-recommender-system/notebook/BX-Users.csv',
    sep=';',
    encoding='latin-1',
    on_bad_lines='skip'
)
```

```python
users.head()
```

```python
# Lets remane columns name
users.rename(columns={
    "User-ID": "user_id",
    "Location": "location",
    "Age": "age",
}, inplace=True)
```

```python
#importing BX-Book-Ratings.csv
ratings = pd.read_csv(
    '/Users/dishitasingh/Desktop/PersonalProjects/end-to-end-recommender-system/notebook/BX-Book-Ratings.csv',
    sep=';',
    encoding='latin-1',
    on_bad_lines='skip'
)
```

```python
ratings.head()
```

```python
# Lets remane columns name
ratings.rename(columns={
    "User-ID": "user_id",
    "Book-Rating": "rating",
}, inplace=True)
```

```python
# Lets store users who had at least rated more than 200 books
x = ratings['user_id']. value_counts() > 200
```

```python
# Now join ratings with books
ratings_with_books = ratings.merge(books, on= 'ISBN')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby("title")['rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
#merge the created number_rating with ratings_with_books

final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
# Lets take those books which got at least 50 rating of user
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
#convert to CSR matrix for easier training

from scipy.sparse import csr_matrix 
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

### Model Training

```python
from sklearn. neighbors import NearestNeighbors
model = NearestNeighbors (algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

### testing model

```python
def recommend_book(book_name):
    # get index of the book
    book_id = np.where(book_pivot.index == book_name)[0][0]
    
    # get nearest neighbors
    distance, suggestion = model.kneighbors(
        book_pivot.iloc[book_id, :].values.reshape(1, -1),
        n_neighbors=6
    )
    
    # print results
    print(f"You searched: {book_name}\n")
    print("The suggestion books are:\n")
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j != book_name:
                print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
